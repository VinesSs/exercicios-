```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data = {
    'Jogador': ['Messi', 'Ronaldo', 'Neymar', 'Messi', 'Ronaldo', 'Neymar'],
    'Ano': [2020, 2020, 2020, 2021, 2021, 2021],
    'Gols': [25, 30, 15, 27, 29, 18]
}
df = pd.DataFrame(data)
jogadores = ['Messi', 'Ronaldo', 'Neymar']
df_filtered = df[df['Jogador'].isin(jogadores)]
df_grouped = df_filtered.groupby(['Jogador', 'Ano']).sum().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='Jogador', y='Gols', hue='Ano', data=df_grouped, palette='viridis')
plt.title('Número de Gols por Jogador em Diferentes Anos')
plt.xlabel('Jogador')
plt.ylabel('Número de Gols')
plt.legend(title='Ano')
plt.show()
```
